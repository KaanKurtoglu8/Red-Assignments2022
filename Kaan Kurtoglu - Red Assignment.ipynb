```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)

data = pd.read_excel("4- BOBİ FRS Nakit Akış Tablosu - Dolaylı Yöntem (Konsolide).xlsx",
                     sheet_name="BOBİ FRS NAT Dolaylı Konsolide",)   
```

```python
columns = list(data.columns)
ACIKLAMALAR = []

for column in columns:   # Iterating every element to find extract the texts that we are going to use in our table
    for i in range(2, len(data)):
        if not pd.isnull(data[column][i]) and data[column][i] != 0:
            ACIKLAMALAR.append(data[column][i])

CARI_DONEM = np.zeros(len(ACIKLAMALAR), dtype=int)   # Filling these 2 columns with zeros
ONCEKI_DONEM = np.zeros(len(ACIKLAMALAR), dtype=int)
```

```python
# Creating a dataframe, representing our table
final_table = pd.DataFrame(list(zip(ACIKLAMALAR, CARI_DONEM, ONCEKI_DONEM)),
               columns =['ACIKLAMALAR', 'CARI_DONEM', 'ONCEKI_DONEM'])

final_table
```
